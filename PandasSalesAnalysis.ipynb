```python
#Data Manipulation
import pandas as pd

#Data Visualization
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#What is the overall sales trend?
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# Grouping month year
df_trend = df.groupby('month_year')['sales'].sum().reset_index()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'],df_trend['sales'], color='#b80045')
plt.xticks(rotation ='vertical',size=8)
plt.show()
```

```python
#Which are the top 10 products by sale?
```

```python
prod_sales = pd.DataFrame(df.groupby('product_name')['sales'].sum())
```

```python
prod_sales=prod_sales.sort_values('sales',ascending = False)
```

```python
#Top 10 products by sales
prod_sales[:10]
```

```python
#Which are the most selling products?
```

```python
most_sell_prod = pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['quantity'])
```

```python
most_sell_prod = most_sell_prod.sort_values('quantity',ascending = False)
```

```python
most_sell_prod[:10]
```

```python
#Which is the most preferred ship Mode?
```

```python
sns.countplot(df['ship_mode'])
plt.figure(figsize=(10,8.5))
plt.show()
```

```python
#Which are the most profitable categories and sub-categories?
```

```python
#Grouping category and subcategory
cat_subcat_profit = pd.DataFrame(df.groupby(['category', 'sub_category']).sum(numeric_only=True)['profit'])
```

```python
#sorting the values
cat_subcat_profit.sort_values(['category','profit'], ascending= False)
```
